```python
import pandas as pd
import os
import shutil
from moexalgo import Ticker, Market
from datetime import datetime, timedelta
from tqdm.notebook import tqdm
```

```python
stocks = Market("shares")
all_stocks = pd.DataFrame(stocks.tickers())
all_stocks
```

```python
all_stocks.to_csv('all_stocks.csv')
```

```python
def load_stock_data(stock_symbol, start_date, end_date):
    df = Ticker(stock_symbol).candles(start=start_date, end=end_date, period='1h')
    print(stock_symbol, 
          ' - Downloading data from', start_date.strftime('%Y-%m-%d'), 'to', end_date.strftime('%Y-%m-%d'), 
          "\n", "Downloaded data from", df['begin'].min().strftime('%Y-%m-%d'), 'to', df['begin'].max().strftime('%Y-%m-%d'), '.')
    return df
```

```python
def update_stock_data(stock_symbol):
    filename = f"{stock_symbol}.csv"

    stock_data = load_stock_data(stock_symbol, datetime(2020, 1, 1), datetime(2020, 12, 30))

    stock_data.to_csv(filename)
```

```python
for ticker in all_stocks["ticker"]:
    try:
        update_stock_data(ticker)
    except Exception as e:
        pass
```
